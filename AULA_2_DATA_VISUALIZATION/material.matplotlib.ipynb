```python
#importando as bibliotecas que iremos utilizar
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

<font color=blue><b> Pyplot </b></font><br>
O Pyplot é um módulo do matplotlib que possui todas as funções para construção do gráfico. 

<font color=blue><b> Subplots </b></font><br>
usando subplots você pode dividir a apresentação dos gráficos em vários gráficos alinhados

```python
#comandos básicos
fig = plt.figure()  # figura vazia sem nenhum eixo
fig = plt.subplots(2, 2)  # figura com vários grids de eixos
```

```python
# Plotando gráficos com subplot:
# criando uma base de dados
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

# Cria uma única figura com um único gráfico usando subplot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Único gráfico')

# Cria dois gráticos e carrega os os dados de uma só vez
# nesse exemplo o eixo y é compartilhado através do parâmetro sharey=True
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, y)
ax1.set_title('Comptartilhando Y')
ax2.scatter(x, y)
```

```python
# Compartilha o eixo X com cada coluna dos subplots
plt.subplots(2, 2, sharex='col')

# Comptatilha o eixo Y com cada linha dos subplots
plt.subplots(2, 2, sharey='row')

# Compartilha ambos X e Y com todos os subplots
plt.subplots(2, 2, sharex='all', sharey='all')

# Note que esta construção e a anterior são iguais
plt.subplots(2, 2, sharex=True, sharey=True)

# Cria uma figura de número 10 com um único subplot
# e limpa todo seu conteúdo
fig, ax=plt.subplots(num=10, clear=True)
```

```python
# mais um exemplo usando dois subplots
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title('Tabela com dois subplots')
plt.ylabel('Oscilação Amortecida')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.xlabel('tempo (s)')
plt.ylabel('Sem amortecimento')

plt.show()
```

```python
#adicionando detalhes ao gráfico
x = np.linspace(0, 2, 100)

plt.title("Nosso Gráfico")
plt.xlabel('eixo x')
plt.ylabel('eixo y')

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')
plt.legend()

plt.show()
```

```python
#reposicionando a Legenda
#best, upper right, upper left, lower left, lower right, right
#center left, center right, lower center, upper center, center
    
x = np.linspace(0, 2, 100)
y = np.linspace(0, 2, 100)

plt.title("Nosso Gráfico")
plt.xlabel('eixo x')
plt.ylabel('eixo y')

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')
plt.legend(loc='best')

plt.show()
```

```python
#fazendo anotações

x = np.linspace(0, 2, 100)
y = np.linspace(0, 2, 100)

plt.plot(x, y,    label='linear')
plt.plot(x, y**2, label='quadratic')
plt.plot(x, y**3, label='cubic')

plt.xlabel('eixo x')
plt.ylabel('eixo y')

plt.title("Nosso Gráfico")

plt.legend()

plt.annotate('convergência', 
             xy=(1, 1), 
             xycoords='data',
             xytext=(0.5, 0.5), 
             textcoords='axes fraction',
             arrowprops=dict(facecolor='black', 
                             shrink=0.05),
             horizontalalignment='right', 
             verticalalignment='top',
            )

plt.show()
```

```python
#colocanco marcadores
x = np.linspace(0, 2, 100)
y = np.linspace(0, 2, 100)

plt.title("Nosso Gráfico")
plt.xlabel('eixo x')
plt.ylabel('eixo y')

plt.plot(x, x, label='linear', marker='o', markersize=4)
plt.plot(x, x**2, label='quadratic', marker='^', markersize=4)
plt.plot(x, x**3, label='cubic')

plt.legend()
plt.show()
```

```python
#colocanco marcadores
#color='green', marker='o', linestyle='dashed', linewidth=2, markersize=12

x = np.linspace(0, 2, 10)
y = np.linspace(0, 2, 10)

plt.title("Nosso Gráfico")
plt.xlabel('eixo x')
plt.ylabel('eixo y')

plt.plot(x, x, label='linear', marker='o', markersize=4)
plt.plot(x, x**2, label='quadratic', marker='^', markersize=4)
plt.plot(x, x**3, color='r', label='cubic', linestyle='dashed')

plt.legend()
plt.show()
```

```python
#salvando o gráfico como uma figura
x = np.linspace(0, 2, 10)
y = np.linspace(0, 2, 10)

plt.title("Nosso Gráfico")
plt.xlabel('eixo x')
plt.ylabel('eixo y')

plt.plot(x, x, label='linear', marker='o', markersize=4)
plt.plot(x, x**2, label='quadratic', marker='^', markersize=4)
plt.plot(x, x**3, color='r', label='cubic', linestyle='dashed')

plt.legend()
plt.savefig('data science aspirantes.png')
plt.show()
```

<font color=blue><b> Outros tipos de gráficos </b></font><br>

```python
#gráfico tipo barras
objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
y_pos = np.arange(len(objects))
performance = [10,8,6,4,2,1]

plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Usage')
plt.title('Programming language usage')

plt.show()
```

```python
#gráfico barras horizontais
objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
y_pos = np.arange(len(objects))
performance = [10,8,6,4,2,1]

plt.barh(y_pos, performance, align='center', alpha=0.5)
plt.yticks(y_pos, objects)
plt.xlabel('Usage')
plt.title('Programming language usage')

plt.show()
```
